## Using Loop

```python
fact =1
n = 5
for i in range(1, n+1):
  fact = fact * i
  print(fact)
```

## Using Recursion

```python
n =4
def factorial(n):
  if n ==0 or n==1:
    return 1

  return n * factorial(n-1)


fact = factorial(n)
fact
```

## example 2 -  **Fibonacci series**

```python
def fib(n):
  if n ==1 or n==2:
    return 1

  return fib (n-1) +fib(n-2)

a = fib(5)
a
```
